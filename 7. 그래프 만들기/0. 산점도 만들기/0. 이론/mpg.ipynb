```python
import pandas as pd
import seaborn as sns

mpg = pd.read_csv('mpg.csv')
```

```python
# sns.scatterplot() : 산점도 만드는 함수
# data : 사용할 데이터
# x, y 지정

# x : displ(배기량), y : hwy(고속도로 연비)
sns.scatterplot(data=mpg, x='displ', y='hwy')
```

```python
sns.scatterplot(data=mpg, x='displ', y='hwy')\
.set(xlim=[3, 6]) # x축 범위 3 ~ 6으로 제한
```

```python
sns.scatterplot(data=mpg, x='displ', y='hwy')\
.set(xlim=[3, 6], ylim=[10, 30]) # y축 범위 10 ~ 30으로 제한
```

```python
# drv별로 표식 색상 다르게 표현
sns.scatterplot(data=mpg, x='displ', y='hwy', hue='drv')
```

```python
# 설명 메시지 숨기기
# 그래프 출력 코드 뒤 ; 입력하면 설명 메시지 출력되지 않음
sns.scatterplot(data=mpg, x='displ', y='hwy', hue='drv');
```
