```python
import pandas as pd
import numpy as np
import seaborn as sns

economics = pd.read_csv('economics.csv')
```

```python
# Q1

# 날짜 시간 타입 변수 만들기
economics['date2'] = pd.to_datetime(economics['date'])

# 연도 변수 만들기
economics['year'] = economics['date2'].dt.year

# x축에 연도 표시
sns.lineplot(data=economics, x='year', y='psavert', errorbar=None)
```

```python
# Q2

# 월 변수 추가
economics['month'] = economics['date2'].dt.month

# 2014년 월별(year 변수는 Q1에서 만듦)
date_2014 = economics.query('year == 2014')

sns.lineplot(data=date_2014, x='month', y='psavert', errorbar=None)
```
