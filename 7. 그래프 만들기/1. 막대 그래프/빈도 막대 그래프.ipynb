```python
import pandas as pd
import seaborn as sns

mpg = pd.read_csv('mpg.csv')
```

```python
# 집단별 빈도표 만들기
# 구동 방식 별 빈도
df_mpg = mpg.groupby('drv', as_index=False)\
.agg(n = ('drv', 'count'))

df_mpg
```

```python
# 그래프 만들기
sns.barplot(data=df_mpg, x='drv', y='n')
```

```python
# sns.countlot()으로 빈도 막대 그래프 바로 만들기
sns.countplot(data=mpg, x='drv')
```

```python
# 값의 고유값 출력 : unique()
mpg['drv'].unique()
```

```python
df_mpg['drv'].unique()
```

```python
# 값 정렬하기 : order 사용
# 4, f, r 순으로 막대 정렬
sns.countplot(data=mpg, x='drv', order=['4', 'f', 'r'])
```

```python
# drv 빈도 높은 순으로 막대 정렬
sns.countplot(data=mpg, x='drv', 
              order=mpg['drv'].value_counts().index)
```
