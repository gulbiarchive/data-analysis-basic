```python
import pandas as pd

result = pd.read_csv('exam.csv')
result
```

```python
# pandas의 오름차순
# df.sort_values('정렬하고 싶은 컬럼명')

# 수학 점수 오름차순
result.sort_values('math')
```

```python
# 수학 점수 내림차순
# df.sort_values('컬럼명', ascending=False)
result.sort_values('math', ascending=False)
```

```python
# nclass, math
result.sort_values(['nclass', 'math'])
```

```python
result.sort_values(['math', 'nclass'], ascending=[False, False])
```

```python
# 반은 오름차순으로 정렬하고 수학 점수는 내림차순으로 정렬
# 단, 반 기준으로 수학 점수를 내림차순한다.

result.sort_values(['nclass', 'math'], ascending=[True, False])
```

```python
# 수학 점수를 기준으로 영어 점수를 오름 차순하고 수학 점수는 내림차순한다.
result.sort_values(['math', 'english'], ascending=[False, True])
```
