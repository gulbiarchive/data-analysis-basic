```python
# 파생 변수 추가하기
'''
total이라는 변수에 math, english, science의 합을 넣어서 추출

행 : ex.query()
열(컬럼) : ex[]
'''
import pandas as pd

ex = pd.read_csv('exam.csv')
ex
```

```python
total = ex['math'] + ex['english'] + ex['science']
total
```

```python
# 파생 변수 추가
# 파일이저장된변수.assign(새 변수명 = 변수를 만드는 공식)
# total = ex['math'] + ex['english'] + ex['science']

ex.assign(total = ex['math'] + ex['english'] + ex['science'])
```

```python
# math, english, science의 평균을 구해서 새로운 컬럼으로 추가해주세요.
# 평균은 실수
# 여러 파생 변수를 한 번에 추가

ex.assign(
    total = ex['math'] + ex['english'] + ex['science'],
    maen = (ex['math'] + ex['english'] + ex['science']) // 3)
```

```python
# test라는 파생 변수를 활용하여 // 과학이 60점 이상인 경우 pass, 미만인 경우 fail
import pandas as pd # 시각
import numpy as np # 수학

ex = pd.read_csv('exam.csv')

'''
df.assign()에 np.where()를 적용하면 
조건에 따라 다른 값을 부여한 변수를 추가할 수 있음
'''
ex.assign(test = np.where(ex['science'] >= 60, 'pass', 'fail'))
```

```python
# test라는 파생변수를 활용하여 수학이 50점 이상인 경우 pass, 미만인 경우 fail
ex.assign(test = np.where(ex['math'] >= 50, 'pass', 'fail'))
```

```python
# 추가한 변수를 pandas 함수와 결합
# 오름차순 -> sort_values()
# total
ex.assign(total = ex['math'] + ex['english'] + ex['science'])\
    .sort_values(['math', 'total'], ascending=[True, False])
```

```python
# 정수인 평균을 내림차순 정렬
ex.assign(mean = ((ex['math'] + ex['english'] + ex['science']) // 3))\
    .sort_values('mean', ascending=False)
```
